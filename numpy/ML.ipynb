```python
import numpy as np

# Coefficient matrix A
A = np.array([[1, 1],
              [2, 2]
              ])

# Constants vector b
b = np.array([2, 5])

# Solve for x (variables vector)
solution = np.linalg.solve(A, b)

print("Solution (x, y,z ):", solution)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

# 1. Generate random data (with some noise)
np.random.seed(42)
x = np.linspace(-10, 10, 30)
y = 0.5 * x**3 - 2 * x**2 + x + 5 + np.random.normal(0, 30, x.shape)

# 2. Fit polynomials
deg1 = np.polyfit(x, y, 1)  # Linear
deg2 = np.polyfit(x, y, 2)  # Quadratic
deg3 = np.polyfit(x, y, 3)  # Cubic (True underlying shape)

# Create polynomial functions
p1 = np.poly1d(deg1)
p2 = np.poly1d(deg2)
p3 = np.poly1d(deg3)

# Generate smooth x for plotting curves
x_fit = np.linspace(-10, 10, 100)

# 3. Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='black', label='Noisy Data', alpha=0.7)
plt.plot(x_fit, p1(x_fit), label='Linear Fit (Degree 1)', linestyle='--', color='red')
plt.plot(x_fit, p2(x_fit), label='Quadratic Fit (Degree 2)', linestyle='-.', color='green')
plt.plot(x_fit, p3(x_fit), label='Cubic Fit (Degree 3)', linestyle='-', color='blue')

plt.title("Polynomial Curve Fitting: Linear vs Quadratic vs Cubic")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
